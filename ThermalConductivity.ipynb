# Удельная теплоёмкость твёрдого тела

## Начальные параметры

В качестве твёрдого тела возьмём медь. Исходные данные: $\rho = 8,92\ \frac{\text{г}}{\text{см}^3}$, $M = 64\ \frac{\text{г}}{\text{моль}}$, $\upsilon = 3,6\cdot10^5\ \frac{\text{см}}{\text{с}}$. Физические константы: $k = 1,38\cdot10^{-16}\ \frac{\text{эрг}}{\text{К}}$, $\hslash = 1,054\cdot10^{-27}\ \text{эрг}\cdot\text{с}$, $N_A = 6,02\cdot10^{23}\ \text{моль}^{-1}$, $R = 8,31\ \frac{\text{Дж}}{\text{моль}\cdot\text{К}}$.

```python
rho = 8.92
M = 64
speed_of_sound = 3.6e5
k = 1.38e-16
h_bar = 1.054e-27
Na = 6.02e23
R = 8.31
```

Рассчитаем концентрацию атомов и температуру Дебая

```python
import math

n = rho*Na/M
ThetaD = h_bar*speed_of_sound*math.pow(6*math.pi*math.pi*n, 1/3)/k

print(f'Концентрация атомов: {n} см^(-3), температура Дебая: {ThetaD:.2f} К')
```

## Построение графика теплоёмкости

Теплоёмкость рассчитывается по формуле:
$$
C_V(T) = 36R\frac{T^3}{\Theta_D^3}\left(\int_0^{\frac{\Theta_D}{T}}\!\frac{x^3dx}{e^x-1} - \frac{\Theta_D^4}{4T^4\left(e^{\frac{\Theta_D}{T}} - 1\right)}\right)
$$

```python
import functools
import warnings

import matplotlib.pyplot as plt
from scipy import integrate
import numpy as np


warnings.filterwarnings("ignore")

NPoints = 5000
temperatures = np.linspace(1e-5, 1350, NPoints)

def under_int_function(x):
    return x*x*x/(np.exp(x) - 1)

def general_conductivity(t, theta, r):
    int_value = integrate.romberg(under_int_function, 1e-5, theta/t)
    return 36*r*math.pow(t/theta, 3)*(int_value - math.pow(theta, 4)/(4*math.pow(t, 4)*(np.exp(theta/t) - 1)))

thermal_conductivity = np.vectorize(functools.partial(general_conductivity, theta=ThetaD, r=R))

conductivities = thermal_conductivity(temperatures)


fig, ax = plt.subplots(figsize=(16,10))
ax.vlines(x=ThetaD, ymin=0, ymax=thermal_conductivity(ThetaD), color='firebrick', alpha=0.7, linewidth=2, linestyle="--")
ax.scatter(x=ThetaD, y=thermal_conductivity(ThetaD), s=75, color='firebrick', alpha=0.7)
ax.plot(temperatures, conductivities, label="Теплоёмкость")
ax.axhline(3*R, linestyle="--")

ax.set_xlabel("T, К")
ax.set_ylabel("C, Дж/(моль·К)")
ax.set_ylim(0, 26)
ax.set_xlim(0, 1400)

new_xticks = np.append(ax.get_xticks(), ThetaD)
ax.set_xticks(new_xticks)

plt.show()
```
